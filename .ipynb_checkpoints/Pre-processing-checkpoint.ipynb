```python
droped_train["Sex"] = training_data["Sex"].map({"Female": 0, "Male": 1})
droped_train["Pregnancy status"] = training_data["Pregnancy status"].map({False: 0, True: 1})
```

```python
import random

def similar_replace_h_w(df, pregnant):
    # Iterate over the DataFrame to replace null values where Pregnancy is False
    
    for index, row in df.iterrows():
        if pregnant == False:
            mask = not row['Pregnancy status']
        else:
            mask = row['Pregnancy status']
        if mask and (pd.isnull(row['Height']) or pd.isnull(row['Weight'])):
            # Ensure Age is not null for the current row
            if pd.notnull(row['Age']):
                # Find similar records: same sex and exact age, excluding the current record
                
                if pregnant:
                    random_add = random.randint(0,20)
                    similar_mask = (df['Sex'] == row['Sex']) & \
                               (df['Age'] == "Adolescent") & \
                               (df.index != index)
                else:
                    random_add = 0
                    similar_mask = (df['Sex'] == row['Sex']) & \
                               (df['Age'] == row['Age']) & \
                               (df.index != index)
                # Calculate replacement values as the mean of similar records
                height_replacement = df.loc[similar_mask, 'Height'].mean() if pd.isnull(row['Height']) else row['Height']
                weight_replacement = df.loc[similar_mask, 'Weight'].mean() if pd.isnull(row['Weight']) else row['Weight']

                # Replace null values with the calculated replacements
                df.at[index, 'Height'] = height_replacement + random_add
                df.at[index, 'Weight'] = weight_replacement + random_add
                

    return df
```

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

def fill_null_ages(df):
    
    temp = df.copy()

    temp['Height_normalized'] = (df['Height'] - df['Height'].mean()) / df['Height'].std()
    temp['Weight_normalized'] = (df['Weight'] - df['Weight'].mean()) / df['Weight'].std()

    n_clusters = 4

    kmeans = KMeans(n_clusters=n_clusters)
    temp['Cluster'] = kmeans.fit_predict(temp[['Height_normalized', 'Weight_normalized']])

    temp['Age'] = pd.to_numeric(df['Age'], errors='coerce')

    cluster_ages = temp.groupby('Cluster')['Age'].mean().to_dict()  # or .mean()

    df['Age'] = temp.apply(lambda row: cluster_ages[row['Cluster']] if np.isnan(row['Age']) else row['Age'], axis=1)
    
    return df
```

```python
new_dataset = similar_replace_h_w(droped_train, False)
```

```python
import random

def replace_pregnant_ages(df):
    mask = df["Pregnancy status"] == 1
    random_states = np.random.rand(mask.sum())

    df.loc[mask, "Age"] = np.where(random_states > 0.33, "Adolescent", "Young Adult")
    
    return df

new_dataset = replace_pregnant_ages(new_dataset)
print(new_dataset.info())
```

```python
new_dataset = similar_replace_h_w(droped_train, True)
print(new_dataset.info())
```

```python
#Extract the age group and replace with the (approximate) number of months for the middle of the age group.

def age_float_conversion(age):
    if age == "Neonate":
        return 0.5
    elif age == "Infant":
        return 6
    elif age == "Child":
        return 6 * 12
    elif age == "Adolescent":
        return 15 * 12
    elif age == "Young Adult":
        return 20 * 12
    else:
        # Assuming 'age' can be directly converted to float if it's not one of the categories above
        try:
            return float(age)
        except ValueError:
            return float('nan')
```

```python
new_dataset['Age'] = new_dataset['Age'].apply(age_float_conversion)
print(new_dataset.info())

fill_ages = new_dataset.copy()
new_dataset = fill_null_ages(fill_ages)
```

```python
new_dataset.info()
```

```python
default_age_numbers = [0.5, 6., 72., 180., 240.]
for i, row in new_dataset.iterrows():
    dist = default_age_numbers
    if row["Age"] not in default_age_numbers:
        dist = [abs(x - row["Age"]) for x in dist]
        minimum = dist.index(min(dist))
        print(default_age_numbers[minimum])
        new_dataset.at[i, 'Age'] = default_age_numbers[minimum]
    
print(new_dataset["Age"].unique())
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

def cramers_v(chi2, n, k1, k2):
    """Calculate Cramér's V statistic for a given Chi2 and sample size."""
    return np.sqrt(chi2 / n / min(k1 - 1, k2 - 1))

# Assuming df is your DataFrame
# Convert 'Pregnancy status' to a string to treat it as a categorical variable
new_dataset['Pregnancy status'] = new_dataset['Pregnancy status'].astype(str)

# Chi-Square Test for Pregnancy status vs. Murmur
contingency_table_murmur = pd.crosstab(new_dataset['Pregnancy status'], new_dataset['Murmur'])
chi2_murmur, p_murmur, _, _ = chi2_contingency(contingency_table_murmur)
cramers_v_murmur = cramers_v(chi2_murmur, new_dataset.shape[0], len(contingency_table_murmur.index), len(contingency_table_murmur.columns))

# Chi-Square Test for Pregnancy status vs. Outcome
contingency_table_outcome = pd.crosstab(new_dataset['Pregnancy status'], new_dataset['Outcome'])
chi2_outcome, p_outcome, _, _ = chi2_contingency(contingency_table_outcome)
cramers_v_outcome = cramers_v(chi2_outcome, new_dataset.shape[0], len(contingency_table_outcome.index), len(contingency_table_outcome.columns))

print(f"Pregnancy status vs. Murmur: Chi2={chi2_murmur}, p={p_murmur}, Cramér's V={cramers_v_murmur}")
print(f"Pregnancy status vs. Outcome: Chi2={chi2_outcome}, p={p_outcome}, Cramér's V={cramers_v_outcome}")
```

```python
murmur_dict = {0: 'Present', 1: 'Unknown', 2: 'Absent'}

outcome_dict = {0: 'Abnormal', 1: 'Normal'}

replace_murmur_outcome = new_dataset.copy()
replace_murmur_outcome["Murmur"] = replace_murmur_outcome["Murmur"].map({"Present": 0, "Unknown": 1, "Absent": 2})
replace_murmur_outcome["Outcome"] = replace_murmur_outcome["Outcome"].map({"Abnormal": 0, "Normal": 1})

```
